clean and process it, perform statistical analysis, and generate visual insights to help the

Part I: Randomized Sales Data Generation (NumPy)
1. Initialize Random Seed
2. Generate Sales Data
• Create a NumPy array of random sales data for 12 months (shape = (12, 4)),
where each column represents a product category:
• Electronics, Clothing, Home & Kitchen, Sports.
• Sales values should be between 1000 and 5000 units per month.
• The range should be randomized using the seed.
3. Convert to DataFrame
• Convert the cleaned NumPy array into a Pandas DataFrame with:
i. Columns: Electronics, Clothing, Home & Kitchen, Sports
ii. Index: Months from Jan to Dec

```python
import numpy as np
import pandas as pd

roll_no =8025320084

# 1. Initialize Random Seed (use your 10-digit roll number)
np.random.seed(roll_no % (2**32))   # replace with your roll number

# 2. Generate Sales Data (12 months × 4 categories)
sales_data = np.random.randint(1000, 5001, size=(12, 4))

# 3. Convert to DataFrame
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

df = pd.DataFrame(sales_data, index=months, columns=columns)

# Display DataFrame
print(df)
```


Part II: Data Manipulation and Analysis (Pandas)
1. Display the first 5 rows and summary statistics of the DataFrame.
2. Calculate the total sales per category and total sales per month.
3. Calculate the average sales growth between consecutive months for each category.
4. Add the following columns to the DataFrame:
• Total Sales: Sum of all categories per month.
• Growth Rate: Percentage change in Total Sales from the previous month.
5. Apply a randomized discount:
• If your roll number is even, apply a 10% discount to the Electronics category.
•
If odd, apply a 15% discount to the Clothing category.

```python
# 1. Display first 5 rows and summary statistics
print("First 5 Rows:\n", df.head())
print("\nSummary Statistics:\n", df.describe())


# 2. Total sales per category and per month
category_total = df.sum()
month_total = df.sum(axis=1)

print("\nTotal Sales per Category:\n", category_total)
print("\nTotal Sales per Month:\n", month_total)


# 3. Average sales growth between consecutive months (per category)
growth = df.pct_change()
avg_growth = growth.mean()

print("\nAverage Growth per Category:\n", avg_growth)


# 4. Add new columns
df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100


# 5. Apply discount based on roll number
roll_no = 8025320084   # your roll number

if roll_no % 2 == 0:
    df["Electronics"] = df["Electronics"] * 0.9   # 10% discount
else:
    df["Clothing"] = df["Clothing"] * 0.85        # 15% discount


# Final DataFrame
print("\nUpdated DataFrame:\n", df)
```

Q.2
For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i.
Find element-wise absolute value
ii.
Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each
row.
iii.
Mean, Median and Standard Deviation of flattened array, of each column, and
each row

```python
import numpy as np

# Given array
array = np.array([[1, -2, 3],
                  [-4, 5, -6]])

# i. Element-wise absolute value
abs_array = np.abs(array)
print("Absolute Value:\n", abs_array)


# ii. Percentiles (25th, 50th, 75th)

# Flattened array
print("\nPercentiles (Flattened):", np.percentile(array.flatten(), [25, 50, 75]))

# Column-wise
print("\nPercentiles (Column-wise):\n", np.percentile(array, [25, 50, 75], axis=0))

# Row-wise
print("\nPercentiles (Row-wise):\n", np.percentile(array, [25, 50, 75], axis=1))


# iii. Mean, Median, Standard Deviation

# Flattened
print("\nFlattened:")
print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Std Dev:", np.std(array))

# Column-wise
print("\nColumn-wise:")
print("Mean:", np.mean(array, axis=0))
print("Median:", np.median(array, axis=0))
print("Std Dev:", np.std(array, axis=0))

# Row-wise
print("\nRow-wise:")
print("Mean:", np.mean(array, axis=1))
print("Median:", np.median(array, axis=1))
print("Std Dev:", np.std(array, axis=1))
```

Q.3
For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
i.
Find floor, ceiling and truncated value, rounded values

```python
import numpy as np

# Given array
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

# Floor
floor_vals = np.floor(a)

# Ceiling
ceil_vals = np.ceil(a)

# Truncate
trunc_vals = np.trunc(a)

# Rounded
round_vals = np.round(a)

# Print results
print("Original Array:\n", a)

print("\nFloor Values:\n", floor_vals)
print("\nCeiling Values:\n", ceil_vals)
print("\nTruncated Values:\n", trunc_vals)
print("\nRounded Values:\n", round_vals)
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
# List
lst = [10, 20, 30, 40]

# Positions to swap (example: index 1 and 3)
i = 1
j = 3

print("Original List:", lst)

# Swapping using temporary variable
temp = lst[i]
lst[i] = lst[j]
lst[j] = temp

print("Updated List:", lst)
```

Q.5 Write a Python program to swap two elements in a set by converting it to a list first.

```python
# Original set
s = {10, 20, 30, 40}

print("Original Set:", s)

# Convert set to list
lst = list(s)

# Positions to swap (example: index 0 and 2)
i = 0
j = 2

# Swap using temporary variable
temp = lst[i]
lst[i] = lst[j]
lst[j] = temp

# Convert back to set
s = set(lst)

print("Updated Set:", s)
```
