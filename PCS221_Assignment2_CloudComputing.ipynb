Q2. Write a program in Python to assign subsequent rows to matrix first row elements.

```python

matrix = np.random.randint(1, 100, size=(5, 5))

print("Original Matrix:")
print(matrix)

# Step 2: Assign first row to all rows
matrix[:] = matrix[0]

# Step 3: Print updated matrix
print("\nUpdated Matrix (First row assigned to all rows):")
print(matrix)
```

Q3. Write a program in Python to group similar elements into matrix.

```python
import numpy as np

data = np.random.randint(1, 6, size=20)

print("Original Data:")
print(data)

unique_values = np.unique(data)

matrix = []

for val in unique_values:
    group = data[data == val]
    matrix.append(group.tolist())

print("\nGrouped Matrix:")
for row in matrix:
    print(row)
```

Q5. Write a program in Python to create an n x n square matrix, where all the sub-matrix has the sum of opposite corner elements as even.

```python
def create_even_corner_matrix(n):
    matrix = np.zeros((n, n), dtype=int)
    for i in range(n):
        for j in range(n):
            matrix[i][j] = 2 * (i + j)
    return matrix

n = 4
matrix = create_even_corner_matrix(n)
print("Matrix:\n", matrix)

# Verify opposite corners of full matrix
print("\nTop-left corner:", matrix[0][0])
print("Bottom-right corner:", matrix[n-1][n-1])
print("Sum:", matrix[0][0] + matrix[n-1][n-1], "-> Even?", (matrix[0][0] + matrix[n-1][n-1]) % 2 == 0)
```

Q6 Write a program in Python to perform row-wise element addition in tuple matrix

```python
# Tuple matrix
matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)

# Row-wise addition
row_sums = tuple(sum(row) for row in matrix)

# Output
print("Tuple Matrix:", matrix)
print("Row-wise Sum:", row_sums)
```
